```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json("characters.json")
```

```python
df = df.drop(labels=['ExtraCults', 'Name', 'Clan', 'Tribe', 'BoundSpirits', 'Age', 'Abilities',
                   'ElementalRunes', 'PowerRunes', 'ConditionRunes', 'CoreRunes', 'Attributes',
                    'Equipment', 'InPlay', 'Updates', 'StandardofLiving', 'Role',
                    'HitLocations', 'HitLocationMap', 'MeleeAttacks', 'RangedAttacks',
                    'Income', 'Ransom', 'SkillCategories', 'Advantages', 'RuneSpells',
                    'SpiritMagic', 'LocationForm', 'Lunars', 'CreationSteps', 'Tags',
                    'Notes', 'Movement', 'Skills', 'Powers',
                    'CurrentHP', 'CurrentMP', 'CurrentRP'], axis=1)
```

```python
homeland = df.groupby("Homeland")
homeland_count = homeland.count()
homeland_mean = homeland.mean()
```

```python
sartar = df.where(df.Homeland == "Sartar")
```

```python
pcs = df.where(df.Homeland != "")
pivot = pcs.groupby(["Homeland", "Cult", "Occupation"])
```

```python
pivot_count = pivot.count().where(pivot.count().Strength > 4).dropna()
pivot_mean = pivot.mean().where(pivot_count.Strength > 4).dropna()
```

```python
pivot_count
```

```python
pivot_mean["Count"] = pivot_count["Strength"]
pivot_mean
```

```python
pivot_mean.to_csv("pivot_table.csv")
```

```python
homeland_mean = homeland_mean.where(homeland_count.Strength > 4).dropna()
homeland_mean
```

```python
homeland_mean.plot(kind="barh", figsize=(36, 18), color=["C1","C0", "C2", "C3", "C4", "C5", "C6"],
    title="Homelands", rot=45)
```
